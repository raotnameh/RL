```python
from main import bandits
import numpy as np
import matplotlib.pyplot as plt

seed = np.random.randint(1,10036)
epsi =[0, 0.01, 0.1, 1]
no_bandit = 10
no_iter = 1000
```

# Stationary case

```python
out_s = []
for eps in epsi:
    prob = bandits(no_bandit,eps,no_iter,seed,starting_reward = 0
                   , weighted_aver = True)
    reward, mean, action = prob.forward()
    out_s.append([reward, mean, action])
```

# Non-Stationary case

```python
out_ns = []

for eps in epsi:
    prob = bandits(no_bandit,eps,no_iter,seed,stationary=False,starting_reward = 0
                   , weighted_aver = False)
    reward, mean,action = prob.forward()
    out_ns.append([reward, mean,action])
```

# Plot for stationary case

```python
for i in out_s:
    plt.plot(i[0])
legend_list = []
for i in epsi:
    legend_list.append("eps = "+ str(i))

plt.ylabel('average reward')
plt.xlabel('steps')
plt.legend(legend_list, loc='best')

fig = plt.gcf()
fig.set_size_inches(10.5, 10.5)
# fig.savefig('test2png.png', dpi=100)
for i,eps in enumerate(epsi):
    print("reward after "+ str(no_iter)+" iteration for epsi ,"+ str(eps) +", is", out_s[i][0][-1])
plt.show()
```

# Plot for non-stationary case

```python
for i in out_ns:
    plt.plot(i[0])
legend_list = []
for i in epsi:
    legend_list.append("eps = "+ str(i))

plt.ylabel('average reward')
plt.xlabel('steps')
plt.legend(legend_list, loc='best')

fig = plt.gcf()
fig.set_size_inches(10.5, 10.5)
# fig.savefig('test2png.png', dpi=100)
for i,eps in enumerate(epsi):
    print("reward after "+ str(no_iter)+" iteration for epsi ,"+ str(eps) +", is", out_ns[i][0][-1])
plt.show()
```
