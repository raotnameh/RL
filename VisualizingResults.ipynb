```python
from main import bandits
import numpy as np

no_bandit = 4
no_iter = 10000
seed = np.random.randint(1,10036)
```

```python
import matplotlib.pyplot as plt
```

```python
eps = 0
prob = bandits(no_bandit,eps,no_iter,seed)
reward, mean = prob.forward()
print("learned distribution", mean)
print("reward after last time step ", reward[-1])
plt.plot(reward)
plt.show()
```
